```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
record = pd.read_csv("Downloads/carnivore-spotter-export3.csv")
```

```python
record.head()
```

```python
record.columns
```

```python
record_cols = ['ID','Vocalization','Vocalization_Description', 'Carnivore_Response',
       'Number_of_Young Species', 'Number_of_Adult Species','Eating','Time_Seen','Latitude', 'Longitude', 'Behavior','Time_Submitted',
       'Reaction_Desription','Species', 'On_Leash', 'Number_of_Adults',
       'Number_of_Children','Confidence','Uploaded_Media', 'Dog_Size',
       'General_Comments','Neighborhood','Animal_Features',
       'Conflict_Description','Reaction', 'Conflict','Number_of_Dogs']
```

```python
record.columns = record_cols
```

```python
record.columns
```

```python
record.Time_Seen.sort_values()
```

```python
species_rec = record[record.Species == "Coyote"].copy()
```

```python
pd.set_option('display.max_columns',None)
```

```python
species_rec.sort_values("Time_Seen")
```

```python
species_rec.dtypes
```

```python
species_rec['Time_Seen'] = pd.to_datetime(species_rec.Time_Seen).copy()
```

```python
species_rec.dtypes
```

```python
species_rec.Time_Seen.dt.year.head()
```

```python
species_rec['Year'] = species_rec.Time_Seen.dt.year
```

```python
species_rec.Year.value_counts().sort_index()
```

```python
species_rec.Year.value_counts().sort_index().plot()
```

```python
dfObj = pd.DataFrame(species_rec, columns = ['Latitude','Longitude'])
```

```python
duplicateLocation = dfObj[dfObj.duplicated(['Latitude','Longitude'])]
```

```python
species_rec.Hour.value_counts().sort_index().plot()
```
