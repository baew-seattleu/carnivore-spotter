```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
record = pd.read_csv("Downloads/carnivore-spotter-export3.csv")
```

```python
record_cols = ['ID','Vocalization','Vocalization_Description', 'Carnivore_Response',
       'Number_of_Young Species', 'Number_of_Adult Species','Eating','Time_Seen','Latitude', 'Longitude', 'Behavior','Time_Submitted',
       'Reaction_Desription','Species', 'On_Leash', 'Number_of_Adults',
       'Number_of_Children','Confidence','Uploaded_Media', 'Dog_Size',
       'General_Comments','Neighborhood','Animal_Features',
       'Conflict_Description','Reaction', 'Conflict','Number_of_Dogs']
```

```python
record.columns = record_cols
```

```python
record['Time_Seen'] = pd.to_datetime(record.Time_Seen).copy()
```

```python
record['Month']=record.Time_Seen.dt.month
```

```python
record['Hour']=record.Time_Seen.dt.hour
```

```python
summer = record[(record.Month>=6) & (record.Month<=8)]
```

```python
summer.Species.value_counts().sort_index().plot(kind = 'bar',figsize=(15,5))
```

```python
winter = record[(record.Month==12) | ((record.Month==1)|(record.Month==2))]
```

```python
winter.Species.value_counts().sort_index().plot(kind = 'bar',figsize=(15,5))
```

```python
spring = record[(record.Month>=3)&(record.Month<=5)]
```

```python
spring.Species.value_counts().sort_index().plot(kind = 'bar',figsize=(15,5))
```

```python
fall = record[(record.Month>=9)&(record.Month<=11)]
```

```python
fall.Species.value_counts().sort_index().plot(kind = 'bar',figsize=(15,5))
```
