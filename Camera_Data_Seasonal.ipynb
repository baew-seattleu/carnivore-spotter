```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
record = pd.read_csv("Downloads/camera_results.csv")
```

```python
camera_sites = pd.read_csv("Downloads/sites_spring.csv")
```

```python
record_cols = ['Begin_Time','Species_Name','Common_Name','Actual_Lat','Actual_Lon']
```

```python
record.columns = record_cols
```

```python
record['Begin_Time'] = pd.to_datetime(record.Begin_Time).copy()
```

```python
record['Month']=record.Begin_Time.dt.month
```

```python
record.Common_Name.nunique()
```

```python
camera_sites.nunique()
```

```python
spring = record[(record.Month>=3)&(record.Month<=5)]
```

```python
spring.Common_Name.value_counts().sort_index().plot(kind = 'bar',figsize=(33,10))
```

```python
summer = record[(record.Month>=6) & (record.Month<=8)]
```

```python
summer.Common_Name.value_counts().sort_index().plot(kind = 'bar',figsize=(33,10))
```

```python
fall = record[(record.Month>=9)&(record.Month<=11)]
```

```python
fall.Common_Name.value_counts().sort_index().plot(kind = 'bar',figsize=(33,10))
```

```python
winter = record[(record.Month==12) | ((record.Month==1)|(record.Month==2))]
```

```python
winter.Common_Name.value_counts().sort_index().plot(kind = 'bar',figsize=(33,10))
```
